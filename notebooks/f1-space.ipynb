```python
import pandas as pd
from collections import defaultdict
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
outfile = "../../mainstream/stems_cpp.out"
with open(outfile) as f:
    data = f.read()
data = data.split("====")[1:]
f1s_by_setup = []
for setup in data:
    entries = setup.split("----")[0].strip().split("\n")
    f1s_by_setup.append(entries)
```

```python
for k, f1s in enumerate(f1s_by_setup):
    plt.figure()
    f1s_z = [1. - float(e) for e in f1s if e != "0"]
    f1s = [1. - float(e) for e in f1s]

    x = range(len(f1s))
    y_ = sorted(f1s)

#     y = np.cumsum(y_)
    plt.plot(x, y_, label='original')
    plt.plot(range(len(f1s_z)), sorted(f1s_z), label='zeros removed')
    plt.legend()
    for xx in [50, 500, 1700, 2750, 3002]:
        plt.axvline(xx, ls='dotted', color='red')
    plt.ylim(0, 1)
    plt.savefig("../plots/f1space/stems_cpp_f1_cdf_{}.pdf".format(k))
```
