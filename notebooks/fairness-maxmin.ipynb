```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results, saved = pickle.load(open('../output/streamer/scheduler/18Q2/fairness_results.pkl'))
```

```python
# print(saved)
table = []
for entry in saved:
    run_id = entry[0]
    run_id = run_id.split("-")[1:]
    scheduler = run_id[0]
    budget = int(run_id[1][1:])
    num_apps = entry[1]
#     entry[2] = [1 - x for x in entry[2]]
    row = [scheduler, num_apps, budget] + entry[2:5] + [entry[-1]]
#     print(row)
    table.append(row)
df = pd.DataFrame(table, columns=['scheduler', 'num_apps', 'budget', 'f1s', 'costs', 'configs', 'total_cost'])    
#     print run_id, budget
#     saved = {}
#     for at, l in zip(att, lines):
#         print l
#         saved[l['num_apps']] = {
#             'sharing': l['sharing'],
#             'fps': l['fps'],
#         }
```

```python
df['f1_min'] = df['f1s'].apply(min)
df['f1_max'] = df['f1s'].apply(max)
df['f1_avg'] = df['f1s'].apply(lambda x: sum(x)/len(x))
df['f1_std'] = df['f1s'].apply(np.std)
```
